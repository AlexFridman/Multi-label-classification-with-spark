```python
import json
import time

def parse_data(data: str):
    row = json.loads(data)
    return row

raw_data.count()
```

```python
clean_data = raw_data.map(lambda line: parse_data(line)).map(lambda row: json.dumps(row))
```

```python
test = sc.parallelize(clean_data.take(20000))
test.saveAsTextFile("hdfs:/master:54310/test")
```
