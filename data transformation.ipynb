```python
import json
```

```python
def extract_labels(line: str):
    all_data = json.loads(line)
    return all_data["Hubs"]
```

```python
labels = data.flatMap(lambda line: extract_labels(line)).persist()
```

```python
def transform_data(data: dict):
    res = {}
    res['Author'] = data['Author']
    res['Id'] = data['Number']
    res['Text'] = data['Text']
    res['Comments'] = data['Comments']
    res['CodeComments'] = data['CodeComments']
    res['Hubs'] = data['Hubs']
    res['Name'] = data['Name']
    res['Tags'] = data['Tags']
    return res
```

```python
transfromed = data.map(lambda row: transform_data(json.loads(row)))
```

```python
transfromed.repartition(4).saveAsTextFile('hdfs://master:54310/transformed')
```
